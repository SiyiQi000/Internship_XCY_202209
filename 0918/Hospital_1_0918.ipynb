```python
import pandas as pd
import numpy as np
import jenkspy
```

```python
pd_reader = pd.read_csv("./Hospital_1.csv")
df = pd.DataFrame(pd_reader)
df
```

```python
#Question1:全国来看（全国均值），医院总床位数和人均医院床位数的变化情况：包括数量和增长率
```

```python
mean_sum_ = []
prepare_mean_sum = locals()

for i in range(0,4):
    mean_sum_ = 'mean_sum_'+ str(2017 + i)  #变量命名
    prepare_mean_sum[mean_sum_] = df.iloc[:,i+7]  #赋值
    print(mean_sum_+': '+"%.0f" %np.mean(prepare_mean_sum[mean_sum_]))  #求均值
#增长率计算见excel表“医院床位数_制图”
```

```python
mean_pers_ = []
prepare_mean_pers = locals()

for j in range(0,4):
    mean_pers_ = 'mean_pers_'+ str(2017 + j)  #变量命名
    prepare_mean_pers[mean_pers_] = df.iloc[:,j+17]  #赋值
    print(mean_pers_+': '+"%.2f" %np.mean(prepare_mean_pers[mean_pers_]))  #求均值
#增长率计算见excel表“医院床位数_制图”
```

```python
#Question2:分4大地理区和7大地理区来看的 人均医院床位数的差异（方差）
```

```python
for k in range(0,4):
    a = pd.DataFrame(df.groupby(df["地理分区（4）"])[str(2017+k)+'每千人医院床位数'].var())

a
```

```python
for l in range(0,4):
    b = df.groupby(df["地理分区（4）"])[str(2017+l)+'每千人医院床位数'].count()
print(b)
c = a.merge(b,how='left',on=['地理分区（4）'])
c
```

```python
#Question3:不同组的老龄化率下（自然断点法，高中低三类老龄化率），总的医院床位数 和 人均医院床位数的差异情况（方差）
```

```python
aging_rate_divide = df["老龄化率"]
```

```python
breaks = jenkspy.jenks_breaks(aging_rate_divide,n_classes = 6)
print(breaks)
#其余步骤见excel表“医院床位数_制图”
```
