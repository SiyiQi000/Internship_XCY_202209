#### 表'3.技术产品-7研发实力-1企业重点实验室.xlsx'   ！注意日期可能是只有年份

```python
import pandas as pd
import numpy as np

#表'3.技术产品-7研发实力-1企业重点实验室.xlsx'
result=df2=pd.read_excel(r'...\1.基本信息-1基本信息（补充后）2.xlsx') 
df2=pd.read_excel(r'...\3.技术产品-7研发实力-1企业重点实验室.xlsx')  #打开表,df2
#df2
result2 = result.merge(df2,how='left',on=['公司编码'])  #将两个表合并到左表result,on'公司编码',result2
#result

result2['认定归属年度']=pd.to_datetime(result2['认定归属年度']) #改为时间格式
result2['成立日期']=pd.to_datetime(result2['成立日期']) #改为时间格式
#print(result2.info())

##b2022
result2_b2022 = result2[result2['认定归属年度']<pd.datetime(2022,1,1)] #只取2022年之前
result2_b2022 = result2_b2022[result2_b2022['成立日期']<pd.datetime(2022,1,1)] #只取2022年之前

result2_b2022 = result2_b2022[~result2_b2022['注册状态'].isin(['注销','已告解散'])]
result2_b2022_count = pd.DataFrame(result2_b2022.groupby(result2_b2022["所属市"])['XXX计数项'].count())

##b2021
result2_b2021 = result2[result2['认定归属年度']<pd.datetime(2021,1,1)] #只取2021年之前
result2_b2021 = result2_b2021[result2_b2021['成立日期']<pd.datetime(2021,1,1)] #只取2021年之前

result2_b2021 = result2_b2021[~result2_b2021['注册状态'].isin(['注销','已告解散'])]
result2_b2021_count = pd.DataFrame(result2_b2021.groupby(result2_b2021["所属市"])['XXX计数项'].count())

##b2020
result2_b2020 = result2[result2['认定归属年度']<pd.datetime(2020,1,1)] #只取2020年之前
result2_b2020 = result2_b2020[result2_b2020['成立日期']<pd.datetime(2020,1,1)] #只取2020年之前

result2_b2020 = result2_b2020[~result2_b2020['注册状态'].isin(['注销','已告解散'])]
result2_b2020_count = pd.DataFrame(result2_b2020.groupby(result2_b2020["所属市"])['XXX计数项'].count())

##b2019
result2_b2019 = result2[result2['认定归属年度']<pd.datetime(2019,1,1)] #只取2019年之前
result2_b2019 = result2_b2019[result2_b2019['成立日期']<pd.datetime(2019,1,1)] #只取2019年之前

result2_b2019 = result2_b2019[~result2_b2019['注册状态'].isin(['注销','已告解散'])]
result2_b2019_count = pd.DataFrame(result2_b2019.groupby(result2_b2019["所属市"])['XXX计数项'].count())

##合并年份4个表并输出
result2_final = result2_b2022_count.merge(result2_b2021_count,result2_b2020_count,result2_b2019_count,how='left',on=['所属市'])
result2_final.to_excel("3.技术产品-7研发实力-1企业重点实验室（输出）.xlsx") #输出结果
```
