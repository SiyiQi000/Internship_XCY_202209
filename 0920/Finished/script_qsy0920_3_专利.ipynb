```python
import openpyxl
import xlrd
import pandas as pd
import numpy as np
from  pandas import DataFrame
from openpyxl import load_workbook
xlrd.xlsx.ensure_elementtree_imported(False, None)
xlrd.xlsx.Element_has_iter = True

result=pd.read_excel(r'E:\思懿-已更新数据文件夹2\qsy_1基本信息\地址匹配.xlsx') 
```
