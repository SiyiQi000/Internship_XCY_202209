#### 表'4.经营-1头部企业营收-1A股'

```python
df2=pd.read_excel(r'E:\思懿-已更新数据文件夹2\qsy_4经营\4.经营-1头部企业营收-1A股_fixdate.xlsx')  #打开表,df2

result2 = result.merge(df2,how='left',on=['公司编码'])  #将两个表合并到左表result,on'公司编码',result2
result2['截止日期']=pd.to_datetime(result2['截止日期']) #改为时间格式

##b2022
result2_b2022 = result2[result2['截止日期']<pd.datetime(2022,1,1)] #只取2022年之前

result2_b2022_sum = pd.DataFrame(result2_b2022.groupby(result2_b2022["所属市"])['一、营业总收入'].sum())
result2_b2022_count = pd.DataFrame(result2_b2022.groupby(result2_b2022["所属市"])['一、营业总收入'].count())
result2_b2022_mean = pd.DataFrame(result2_b2022.groupby(result2_b2022["所属市"])['一、营业总收入'].mean())

#改名
result2_b2022_sum=result2_b2022_sum.rename(columns={'一、营业总收入':'一、营业总收入_b2022_sum'})
result2_b2022_count=result2_b2022_count.rename(columns={'一、营业总收入':'一、营业总收入_b2022_count'})
result2_b2022_mean=result2_b2022_mean.rename(columns={'一、营业总收入':'一、营业总收入_b2022_mean'})

tbs = [result2_b2022_count,result2_b2022_mean]
result2_b2022_final = result2_b2022_sum
for tb in tbs:
    result2_b2022_final = result2_b2022_final.merge(tb, on=['所属市'], how='outer')

##b2021
result2_b2021 = result2[result2['截止日期']<pd.datetime(2021,1,1)] #只取2021年之前

result2_b2021_sum = pd.DataFrame(result2_b2021.groupby(result2_b2021["所属市"])['一、营业总收入'].sum())
result2_b2021_count = pd.DataFrame(result2_b2021.groupby(result2_b2021["所属市"])['一、营业总收入'].count())
result2_b2021_mean = pd.DataFrame(result2_b2021.groupby(result2_b2021["所属市"])['一、营业总收入'].mean())

#改名
result2_b2021_sum=result2_b2021_sum.rename(columns={'一、营业总收入':'一、营业总收入_b2021_sum'})
result2_b2021_count=result2_b2021_count.rename(columns={'一、营业总收入':'一、营业总收入_b2021_count'})
result2_b2021_mean=result2_b2021_mean.rename(columns={'一、营业总收入':'一、营业总收入_b2021_mean'})

tbs = [result2_b2021_count,result2_b2021_mean]
result2_b2021_final = result2_b2021_sum
for tb in tbs:
    result2_b2021_final = result2_b2021_final.merge(tb, on=['所属市'], how='outer')

##b2020
result2_b2020 = result2[result2['截止日期']<pd.datetime(2020,1,1)] #只取2020年之前

result2_b2020_sum = pd.DataFrame(result2_b2020.groupby(result2_b2020["所属市"])['一、营业总收入'].sum())
result2_b2020_count = pd.DataFrame(result2_b2020.groupby(result2_b2020["所属市"])['一、营业总收入'].count())
result2_b2020_mean = pd.DataFrame(result2_b2020.groupby(result2_b2020["所属市"])['一、营业总收入'].mean())

#改名
result2_b2020_sum=result2_b2020_sum.rename(columns={'一、营业总收入':'一、营业总收入_b2020_sum'})
result2_b2020_count=result2_b2020_count.rename(columns={'一、营业总收入':'一、营业总收入_b2020_count'})
result2_b2020_mean=result2_b2020_mean.rename(columns={'一、营业总收入':'一、营业总收入_b2020_mean'})

tbs = [result2_b2020_count,result2_b2020_mean]
result2_b2020_final = result2_b2020_sum
for tb in tbs:
    result2_b2020_final = result2_b2020_final.merge(tb, on=['所属市'], how='outer')

##b2019
result2_b2019 = result2[result2['截止日期']<pd.datetime(2019,1,1)] #只取2019年之前

result2_b2019_sum = pd.DataFrame(result2_b2019.groupby(result2_b2019["所属市"])['一、营业总收入'].sum())
result2_b2019_count = pd.DataFrame(result2_b2019.groupby(result2_b2019["所属市"])['一、营业总收入'].count())
result2_b2019_mean = pd.DataFrame(result2_b2019.groupby(result2_b2019["所属市"])['一、营业总收入'].mean())

#改名
result2_b2019_sum=result2_b2019_sum.rename(columns={'一、营业总收入':'一、营业总收入_b2019_sum'})
result2_b2019_count=result2_b2019_count.rename(columns={'一、营业总收入':'一、营业总收入_b2019_count'})
result2_b2019_mean=result2_b2019_mean.rename(columns={'一、营业总收入':'一、营业总收入_b2019_mean'})

tbs = [result2_b2019_count,result2_b2019_mean]
result2_b2019_final = result2_b2019_sum
for tb in tbs:
    result2_b2019_final = result2_b2019_final.merge(tb, on=['所属市'], how='outer')

##合并年份4个表并输出
tbs = [result2_b2021_final,result2_b2020_final,result2_b2019_final]
result2_final = result2_b2022_final
for tb in tbs:
    result2_final = result2_final.merge(tb, on=['所属市'], how='outer')

result2_final.to_excel("4.经营-1头部企业营收-1A股_output.xlsx") #输出结果A
```
